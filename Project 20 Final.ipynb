```python
%matplotlib inline

import numpy
import pandas as pd
import matplotlib as matplot 
import seaborn
```

```python
data = pd.read_csv('C:/Users/swana/Downloads/Crimes.csv',parse_dates={"ArrestDateTime":["DATE.OCC","TIME.OCC"]})
```

```python
data.head()
```

```python
def converttodate():
    new=[]
    toconvert = data["ArrestDateTime"]
    for a in toconvert:
        obj = a.split(" ")   
        time = str(obj[1])
        if len(time)==4:
            time = time[:2]+':'+time[2:]         
            new.append(obj[0]+" "+time)
        elif len(time)==3:
            time = '0'+time[:1]+':'+time[1:]         
            new.append(obj[0]+" "+time)
        elif len(time)==2:        
            time = '0'+time[0]+':'+time[0:]
            new.append(obj[0]+" "+time)        
        else:
            time = time+'0:00'
            new.append(obj[0]+" "+time)
        
    return new
```

```python
newArrestDateTime = converttodate()
data = data.drop("ArrestDateTime", axis=1)
data["ArrestDateTime"] = newArrestDateTime
```
